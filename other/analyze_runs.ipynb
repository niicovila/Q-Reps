```python
lunar_lander = pd.read_csv('/Users/nicolasvila/workplace/uni/tfg_v2/tests/CartPole_qreps_v2-1.csv')
lunar_lander.drop(['timestamp', 
                        'trial_id', 
                        'checkpoint_dir_name', 
                        'done',
                        'training_iteration',
                        'date',
                        'time_this_iter_s',
                        'time_total_s',
                        'pid',
                        'hostname',
                        'node_ip',
                        'time_since_restore',
                        'iterations_since_restore',
                        'experiment_tag',
                        'config/env_id',
                        'config/exp_name',
                        'config/seed',
                        'config/torch_deterministic',
                        'config/cuda',
                        'config/track',
                        'config/wandb_project_name',
                        'config/wandb_entity',
                        'config/capture_video',
                        'config/__trial_index__',
                        ], axis=1, inplace=True)

lunar_lander.reset_index(drop=True, inplace=True)
```

```python
result_df = lunar_lander.groupby(lunar_lander.index // 3).agg({'reward': 'mean', **{col: 'first' for col in lunar_lander.columns if col != 'reward'}}).reset_index(drop=True)
result_df = result_df.sort_values(by='reward', ascending=False)
result_df.reset_index(drop=True, inplace=True)
result_df.drop(['config/total_timesteps',
                        'config/gamma',
                        'config/q_histogram',
                        'config/batch_size',
                        'config/num_iterations',
                        'config/eta',
                        'config/tau',
                        'config/minibatch_size',
                        'config/parametrized_sampler',
                        'config/saddle_point_optimization',
                        'config/target_network_frequency',
                        # 'config/use_kl_loss',
                        'config/target_network'
                        ], axis=1, inplace=True)

result_df.reset_index(drop=True, inplace=True)
result_df.head()
# result_df.to_csv('CartPole_qreps_v2-1-averaging.csv', index=False)
```
