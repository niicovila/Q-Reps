```python
import pandas as pd

# Assuming you have two dataframes: top_10_rewards_cp and top_10_rewards_ll

# Function to find similar configurations
def find_similar_configurations(df1, df2):
    similar_configurations = []

    for idx1, row1 in df1.iterrows():
        for idx2, row2 in df2.iterrows():
            similarity_score = sum(row1[:-1] == row2[:-1])
            if similarity_score > 0:
                similar_configurations.append((idx1, idx2, similarity_score))

    return similar_configurations

# Find similar configurations
similar_configurations = find_similar_configurations(top_10_rewards_cp, top_10_rewards_ll)
similar_configurations.sort(key=lambda x: x[2], reverse=True)
# Print the top similar configurations
for config in similar_configurations[:10]:
    idx1, idx2, similarity_score = config
    print("Similarity Score:", similarity_score)
    print("Configuration CP (Index {}):".format(idx1))
    print(top_10_rewards_cp.iloc[idx1])
    print("Configuration LL (Index {}):".format(idx2))
    print(top_10_rewards_ll.iloc[idx2])
    print("="*50)
```
